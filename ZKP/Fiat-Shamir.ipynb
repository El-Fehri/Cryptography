```python
import hashlib
```

###  Fonction de hachage

```python
def H(x, m):
    x_bytes = Integer(x).digits(base=256)
    data = bytearray(x_bytes[::-1]) + m.encode()
    h = hashlib.sha256(bytes(data)).hexdigest()
    return int(h, 16) % 2
```

### Étape 1 : génération de la preuve

```python
def FSProof(n,m,s):
    while True:
        r = randint(2, n - 1)
        if gcd(r, n) == 1:  # r ∈ Z*_n
            break

    x = power_mod(r, 2, n)
    e = H(x, m)
    y = (r * power_mod(s, e, n)) % n

    return (x,y)
```

### Étape 2 : vérification

```python
def FSVerify(n,v,m,x,y):
    e_verif = H(x, m)
    lhs = power_mod(y, 2, n)
    rhs = (x * power_mod(v, e_verif, n)) % n
    
    return lhs == rhs
```
