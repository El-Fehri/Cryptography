## Init the curve P-256

```python
# a large prime number p
p = 0xffffffff00000001000000000000000000000000ffffffffffffffffffffffff
# the finite field of Z/pZ
K = GF(p)
# equation parameters a and b such as : y^2 = x^3 + ax + b
a = K(0xffffffff00000001000000000000000000000000fffffffffffffffffffffffc)
b = K(0x5ac635d8aa3a93e7b3ebbd55769886bc651d06b0cc53b0f63bce3c3e27d2604b)
# Define the EC 
E = EllipticCurve(K, (a, b))
# the generator of the EC
G = E(0x6b17d1f2e12c4247f8bce6e563a440f277037d812deb33a0f4a13945d898c296, 0x4fe342e2fe1a7f9b8ee7eb4a7c0f9e162bce33576b315ececbb6406837bf51f5)
E.set_order(0xffffffff00000000ffffffffffffffffbce6faada7179e84f3b9cac2fc632551 * 0x1)
```

## Step 1 : Alice and Bob chose secret n and m and calculate points A and B

## Alice 

```python
# ALICE chose a random element from the field
#n = int(K.random_element()) # randomly

# fixed fro brute forcing
n = 0xffff
# Calculate P = nG
A = n*G
A, n
```

## Bob

```python
# BOB chose a random element from the field
# m = int(K.random_element()) # randomly

# fixed for brute forcing
m = 0xfff
# Calculate P = nG
B = m*G
B, m
```

## Step 2 : ALICE and BON exchange result point A and B and each one calculate his secret key multiple point of the other

## ALICE

```python
# already has n and must calculate n*B (B : public key of BOB)
SharedK1 = n*B
SharedK1
```

## BOB

```python
# already has m and must calculate m*A (A : public key of ALICE)
SharedK2 = m*A
SharedK2
```

## Verify 

```python
if SharedK1 == SharedK2 :
    print(f"Alice and Bob have same shared key : {SharedK1}")
else :
    print(f"The Shared Key failed because : {Sharedk1} not equal {SharedK2}")
```
