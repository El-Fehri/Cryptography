# RC4 Stream Cipher Complete Demonstration
## A Pure Stream Cipher Implementation in SageMath

This notebook demonstrates RC4 as a true stream cipher, showing keystream generation separate from data encryption.

## 1. Core RC4 Stream Cipher Implementation

```python
def rc4_ksa(key):
    """RC4 Key Scheduling Algorithm - Initial permutation"""
    if isinstance(key, str):
        key = key.encode()
    
    S = list(range(256))
    j = 0
    
    for i in range(256):
        j = (j + S[i] + key[i % len(key)]) % 256
        S[i], S[j] = S[j], S[i]
    
    return S

def rc4_prga_generator(S):
    """RC4 PRGA as a generator - produces infinite keystream"""
    S = S[:]  # Copy the S-box
    i = 0
    j = 0
    
    while True:
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        yield S[(S[i] + S[j]) % 256]

def rc4_keystream(key, length):
    """Generate specific length of keystream"""
    S = rc4_ksa(key)
    generator = rc4_prga_generator(S)
    return [next(generator) for _ in range(length)]

def rc4_stream_encrypt(key, plaintext):
    """Stream cipher encryption - XOR with keystream"""
    if isinstance(plaintext, str):
        plaintext = plaintext.encode()
    
    keystream = rc4_keystream(key, len(plaintext))
    ciphertext = bytes([p ^ k for p, k in zip(plaintext, keystream)])
    
    return ciphertext, keystream

def rc4_stream_decrypt(key, ciphertext):
    """Stream cipher decryption - identical to encryption"""
    return rc4_stream_encrypt(key, ciphertext)[0]
```

## 2. Stream Cipher Fundamental Properties

```python
print("=" * 70)
print("STREAM CIPHER FUNDAMENTALS")
print("=" * 70)

# Property 1: Keystream independent of plaintext
key = "DemoKey"
keystream1 = rc4_keystream(key, 10)
keystream2 = rc4_keystream(key, 10)
print(f"\n1. Same key -> Same keystream: {keystream1 == keystream2}")

# Property 2: Different keys -> Different keystream
keystream3 = rc4_keystream("DifferentKey", 10)
print(f"2. Different keys -> Different keystream: {keystream1 != keystream3}")

# Property 3: Keystream can be generated before seeing plaintext
pre_keystream = rc4_keystream("PreKey", 5)
print(f"3. Keystream pre-generated: {pre_keystream}")

# Property 4: No padding needed for any length
for length in [3, 7, 16, 32]:
    ks = rc4_keystream("Key", length)
    print(f"4. Length {length}: {'âœ“' if len(ks) == length else 'âœ—'}")
```

## 3. Visual Stream Cipher Operation

```python
def visualize_stream_cipher(key, plaintext):
    """Visual representation of stream cipher operation"""
    print("\n" + "â”€" * 80)
    print(f"STREAM CIPHER: Key='{key}', Plaintext='{plaintext}'")
    print("â”€" * 80)
    
    # Generate keystream
    plaintext_bytes = plaintext.encode()
    keystream = rc4_keystream(key, len(plaintext_bytes))
    ciphertext = bytes([p ^ k for p, k in zip(plaintext_bytes, keystream)])
    
    # Create visual table
    print("\nCharacter:   ", end="")
    for c in plaintext:
        print(f"  {c:^8}", end="")
    
    print("\nASCII:       ", end="")
    for b in plaintext_bytes:
        print(f"  {b:^8}", end="")
    
    print("\n" + " " * 12, end="")
    for _ in plaintext:
        print(f"  {chr(0x2295):^8}", end="")  # XOR symbol
    
    print("\nKeystream:   ", end="")
    for k in keystream:
        print(f"  {k:^8}", end="")
    
    print("\n" + " " * 12, end="")
    for _ in plaintext:
        print(f"  {chr(0x2550):^8}", end="")  # Line symbol
    
    print("\nCiphertext:  ", end="")
    for c in ciphertext:
        print(f"  {c:^8}", end="")
    
    print("\nCipher hex:  ", end="")
    hex_str = ciphertext.hex().upper()
    for i in range(0, len(hex_str), 2):
        print(f"  {hex_str[i:i+2]:^8}", end="")
    print("\n" + "â”€" * 80)
    
    return ciphertext, keystream

# Demonstrate with different messages
print("\n=== VISUAL DEMONSTRATION ===")
c1, ks1 = visualize_stream_cipher("Key1", "Hello")
c2, ks2 = visualize_stream_cipher("Key1", "World")
```

## 5. Stream Cipher XOR Operation Deep Dive

```python
def xor_demonstration():
    """Detailed XOR operation demonstration"""
    print("\n" + "â•" * 70)
    print("XOR OPERATION IN GF(256)")
    print("â•" * 70)
    
    try:
        from sage.all import GF
        F = GF(256)
        use_sage = True
    except:
        use_sage = False
        print("(Running without SageMath GF(256) - using regular XOR)")
    
    # Example values
    plaintext_vals = [0x48, 0x65, 0x6C, 0x6C, 0x6F]  # "Hello"
    keystream_vals = [0xB2, 0x9F, 0x51, 0xAC, 0x03]
    
    print("\nMathematical representation:")
    print("    Plaintext:  " + " ".join([f"0x{b:02x}" for b in plaintext_vals]))
    print("    Keystream:   " + " ".join([f"0x{b:02x}" for b in keystream_vals]))
    print("    " + "â”€" * 35)
    
    # XOR in regular integer arithmetic
    xor_regular = [p ^ k for p, k in zip(plaintext_vals, keystream_vals)]
    print("    XOR (int):   " + " ".join([f"0x{b:02x}" for b in xor_regular]))
    
    if use_sage:
        # XOR as addition in GF(256)
        xor_gf256 = []
        for p, k in zip(plaintext_vals, keystream_vals):
            gf_sum = F(p) + F(k)
            xor_gf256.append(gf_sum.integer_form())
        print("    XOR (GF256): " + " ".join([f"0x{b:02x}" for b in xor_gf256]))
        print(f"\nVerification: Regular XOR == GF(256) addition: {xor_regular == xor_gf256}")

xor_demonstration()
```

## 7. Practical Stream Cipher Applications

```python
class RC4Stream:
    """Practical RC4 stream cipher class"""
    
    def __init__(self, key):
        self.key = key
        self.reset()
    
    def reset(self):
        """Reset to initial state"""
        self.S = rc4_ksa(self.key)
        self.generator = rc4_prga_generator(self.S)
    
    def encrypt_stream(self, data):
        """Encrypt streaming data"""
        if isinstance(data, str):
            data = data.encode()
        
        result = bytearray()
        for byte in data:
            keystream_byte = next(self.generator)
            result.append(byte ^ keystream_byte)
        
        return bytes(result)
    
    def decrypt_stream(self, data):
        """Decrypt streaming data"""
        return self.encrypt_stream(data)

# Demonstrate streaming capability
print("\n" + "=" * 70)
print("STREAMING DATA DEMONSTRATION")
print("=" * 70)

# Create stream cipher instance
rc4_stream = RC4Stream("StreamKey")

# Simulate streaming data in chunks
chunks = ["First chunk ", "second chunk ", "third chunk"]
encrypted_chunks = []

print("\nEncrypting streaming data:")
for i, chunk in enumerate(chunks):
    encrypted = rc4_stream.encrypt_stream(chunk)
    encrypted_chunks.append(encrypted)
    print(f"Chunk {i+1}: '{chunk}' -> {encrypted.hex()[:20]}...")

# Reset for decryption
rc4_stream.reset()
print("\nDecrypting streaming data:")
for i, encrypted in enumerate(encrypted_chunks):
    decrypted = rc4_stream.decrypt_stream(encrypted)
    print(f"Chunk {i+1}: {encrypted.hex()[:20]}... -> '{decrypted.decode()}'")
```

## 8. Performance Analysis

```python
import time

print("\n" + "=" * 70)
print("PERFORMANCE ANALYSIS")
print("=" * 70)

def benchmark_rc4():
    """Benchmark RC4 stream cipher performance"""
    
    sizes = [100, 1000, 10000, 100000]
    results = []
    
    for size in sizes:
        data = b"X" * size
        key = "BenchmarkKey"
        
        # Benchmark keystream generation
        start = time.time()
        keystream = rc4_keystream(key, size)
        ks_time = time.time() - start
        
        # Benchmark encryption
        start = time.time()
        ciphertext, _ = rc4_stream_encrypt(key, data)
        enc_time = time.time() - start
        
        results.append({
            'size': size,
            'ks_time': ks_time,
            'enc_time': enc_time,
            'ks_speed': size / ks_time / 1024 / 1024 if ks_time > 0 else 0,  # MB/s
            'enc_speed': size / enc_time / 1024 / 1024 if enc_time > 0 else 0  # MB/s
        })
    
    # Display results
    print("\n{:<12} {:<20} {:<20} {:<20} {:<20}".format(
        "Size", "Keystream Time(s)", "Encrypt Time(s)", "KS Speed(MB/s)", "Enc Speed(MB/s)"))
    print("-" * 92)
    
    for r in results:
        print("{:<12} {:<20.6f} {:<20.6f} {:<20.2f} {:<20.2f}".format(
            r['size'], r['ks_time'], r['enc_time'], r['ks_speed'], r['enc_speed']))

benchmark_rc4()
```

## 10. Interactive Stream Cipher Demo

```python
def interactive_rc4_demo():
    """Interactive demonstration of RC4 stream cipher"""
    
    print("\n" + "ðŸŽ¯" * 35)
    print("INTERACTIVE RC4 STREAM CIPHER DEMO")
    print("ðŸŽ¯" * 35)
    
    # Get user input
    try:
        key = input("\nEnter encryption key (press Enter for 'DefaultKey'): ") or "DefaultKey"
        plaintext = input("Enter message to encrypt (press Enter for 'Hello RC4'): ") or "Hello RC4"
    except:
        # For non-interactive environments
        key = "DefaultKey"
        plaintext = "Hello RC4"
        print(f"\nUsing default key: '{key}'")
        print(f"Using default message: '{plaintext}'")
    
    print("\n" + "â”€" * 60)
    print("STREAM CIPHER OPERATION:")
    print("â”€" * 60)
    
    # Step 1: Show KSA initialization
    print(f"\nStep 1: Key Scheduling (KSA)")
    print(f"  Key: '{key}'")
    S = rc4_ksa(key)
    print(f"  S-box initialized: {S[:5]}... (first 5 bytes)")
    
    # Step 2: Generate keystream
    print(f"\nStep 2: Keystream Generation (PRGA)")
    keystream = rc4_keystream(key, len(plaintext))
    print(f"  Keystream bytes: {keystream[:5]}... (first 5)")
    print(f"  Keystream hex: {bytes(keystream).hex()[:20]}...")
    
    # Step 3: Encrypt
    print(f"\nStep 3: Encryption (XOR)")
    plaintext_bytes = plaintext.encode()
    ciphertext = bytes([p ^ k for p, k in zip(plaintext_bytes, keystream)])
    print(f"  Plaintext:  {plaintext_bytes.hex()[:20]}...")
    print(f"  Keystream:  {bytes(keystream).hex()[:20]}...")
    print(f"  Ciphertext: {ciphertext.hex()[:20]}...")
    
    # Step 4: Decrypt
    print(f"\nStep 4: Decryption (same XOR)")
    decrypt_keystream = rc4_keystream(key, len(ciphertext))
    decrypted = bytes([c ^ k for c, k in zip(ciphertext, decrypt_keystream)])
    print(f"  Ciphertext: {ciphertext.hex()[:20]}...")
    print(f"  Keystream:  {bytes(decrypt_keystream).hex()[:20]}...")
    print(f"  Decrypted:  {decrypted.decode()}")
    
    print("\n" + "â”€" * 60)
    print(f"âœ… Success! Message: '{decrypted.decode()}'")

# Run interactive demo
interactive_rc4_demo()
```
